```python
SILVER_TABLE = "tbl_silver_part"
GOLD_TABLE   = "tbl_gold_amount_per_event_type"


# query def

query = f"""
SELECT 
    DATE(timestamp)  AS event_date,
    symbol,
    exchange,
    event_type,
    COUNT(*)         AS total_events
FROM {SILVER_TABLE}
WHERE DATE(timestamp) = current_date()
GROUP BY 
    DATE(timestamp), 
    symbol, 
    exchange,
    event_type
ORDER BY symbol, exchange, event_type DESC;
"""

df_gold = spark.sql(query)

#write to gold

(df_gold.write
    .format("delta")
    .mode("overwrite")
    .option("overwriteSchema", "true")
    .partitionBy("event_date")          # partyc po dacie
    .saveAsTable(GOLD_TABLE))
```
